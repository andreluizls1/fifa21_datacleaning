##1.1. Breves conclusões e identificação dos objetivos concretos do projeto.

- As colunas 'LongName', 'photoUrl', 'playerUrl', podem ser removidas.
- A coluna 'Club' pode ser limpa para remover os caracteres \n\n\n\n de seus nomes.
- A coluna 'Contract' está em um formato que pode ser melhorado, com a criação de uma coluna 'Contract_start_year' e 'Contract_end_year' e sua alteração para três status de contrato: 'Active', 'Free' e 'On Loan'.
- As colunas 'Height' e 'Weight' estão com dtype object, e alguns registros estão fora do Sistema Internacional de Unidades, o que deve ser corrigido.
- A coluna 'Joined' pode ser convertido para datetime, e a coluna 'Loan Date End' pode ser transformada em base para preencher o 'Contract_end_year' dos jogadores On Loan.
- As colunas que envolvem valores monetários, devem ser convertidas para o dtype float e ter removidos os símbolos de Euro e multiplicar valores terminados em K por 1.000 e em M por 1.000.000.
- As colunas 'W/F', 'SM' e 'IR' possuem uma estrela que deve ser removida, e os valores transformados no dtype int.
- A coluna Hits tem um problema parecido com o das colunas de valores com um K equivalendo a 1000, com o adendo de um problema com os tipos dos dados.


#2. Limpeza e manipulação dos dados

```python
# Removendo as colunas que não ficarão no dataframe

df = df.drop(['LongName', 'photoUrl', 'playerUrl'], axis = 1)
df.head()
```

```python
# Removendo os caracteres da coluna 'Club'

df['Club'] = df['Club'].str[4:]
df.head()
```

```python
# Criando as colunas 'Contract_start_year' e 'Contract_end_year

df[['Contract_start_year', 'Contract_end_year']] = df['Contract'].str.split(' ~ ', expand = True)
df.head()
```

```python
# Realocando as colunas

coluna_movida = df.pop('Contract_start_year')
df.insert(8, 'Contract_start_year', coluna_movida)
df.head()
```

```python
# Agora a segunda

coluna_movida2 = df.pop('Contract_end_year')
df.insert(9, 'Contract_end_year', coluna_movida2)
df.head()
```

```python
# Alterando o preenchimento de 'Contract'
# Definindo uma função para fazer o mapeamento dos status

def mapear_status(status):
  if 'On Loan' in status:
    return 'On Loan'
  elif 'Free' in status:
    return 'Free'
  else:
    return 'Active'
```

```python
# Aplicando o mapeamento

df['Contract'] = df['Contract'].apply(mapear_status)
df['Contract'].unique()
```

```python
# Definindo uma função para converter a altura
import re

def conversor_altura(altura):
  if 'cm' in altura:
    return round((float(re.search(r'\d+', altura).group())), 2)
  else:
    partes = altura.split("'")
    pes = float(partes[0])
    polegadas = float(partes[1].replace("\"", ""))
    altura_cm = pes * 30.48 + polegadas * 2.54
    return round(altura_cm, 2)
```

```python
df['Height'] = df['Height'].apply(conversor_altura)
df.head()
```

```python
# Definindo uma função para converter peso

def conversor_peso(peso):
  if 'lbs' in peso:
    return round((float(peso.replace('lbs', '').strip()) * 0.453592), 2)
  elif 'kg' in peso:
    return round((float(peso.replace('kg', '').strip())), 2)
  else:
    return float(peso)
```

```python
# Aplicando o conversor de peso

df['Weight'] = df['Weight'].apply(conversor_peso)
df.head()
```

```python
# Agora é o momento de alterar as colunas de contrato
# Vamos primeiro checar os valores únicos nas colunas

print(df['Contract'].unique())
print(df['Contract_start_year'].unique())
print(df['Contract_end_year'].unique())
```

```python
# Para lidar com os atletas 'Free' a solução escolhida foi aplicar 2020 para 'Contract_end_year', ano de lançamento do jogo
# Primeiro vamos buscar na coluna 'Contract_end_year' os atletas 'Free'

idx_free = df[df['Contract_start_year'] == 'Free'].index
idx_free
```

```python
# Agora vamos usar seus index para substituir os valores na coluna 'Contract_end_year'

df.loc[idx_free, 'Contract_end_year'] = df.loc[idx_free].fillna('2020')
df.loc[idx_free].head()
```

```python
# Observando a coluna Joined nessa seleção de atletas, é possível buscar seus 'Contract_start_year'

df.loc[idx_free, 'Joined'].info()
```

```python
# Vamos realocar nesses valores

df.loc[idx_free, 'Contract_start_year'] = df.loc[idx_free, 'Joined']
df.loc[idx_free].head()
```

```python
# Mantendo apenas o ano

df.loc[idx_free, 'Contract_start_year'] = pd.to_datetime(df.loc[idx_free, 'Contract_start_year']).dt.year
df.loc[idx_free].head()
```

```python
# Os atletas 'Free' possuem um erro no preenchimento do clube
# Vamos substituir por 'No Club'

df.loc[idx_free, 'Club'] = df.loc[idx_free, 'Club'].apply(lambda x: x.replace('lub', 'No Club').strip())
df.loc[idx_free].head()
```

```python
# Agora os atletas 'On Loan'

idx_loan = df[df['Contract'] == 'On Loan'].index
idx_loan
```

```python
# Substituindo os valores

df.loc[idx_loan, 'Contract_end_year'] = df.loc[idx_loan, 'Loan Date End']
df.loc[idx_loan].head()
```

```python
# Mantendo apenas os anos nas colunas

df.loc[idx_loan, 'Contract_end_year'] = pd.to_datetime(df.loc[idx_loan, 'Contract_end_year']).dt.year
df.loc[idx_loan].head()
```

```python
# Checando a coluna 'Contract_end_year'

df['Contract_end_year'].unique()
```

```python
# Consertando a coluna Contract_start_year

df['Contract_start_year'].unique()
```

```python
# Checando os indices de 'Loan' novamente

df.loc[idx_loan, 'Contract_start_year'].unique()
```

```python
# Removendo a string ' On Loan'

df.loc[idx_loan, 'Contract_start_year'] = df.loc[idx_loan, 'Contract_start_year'].apply(lambda x: x.replace(' On Loan', '').strip())
df.loc[idx_loan].head()
```

```python
# Convertendo para o formato correto

df.loc[idx_loan, 'Contract_start_year'] = pd.to_datetime(df.loc[idx_loan, 'Contract_start_year']).dt.year
df['Contract_start_year'].unique()
```

```python
# Convertendo toda a coluna 'Contract_start_year'

pd.to_datetime(df['Contract_start_year']).dt.year
```

```python
# Removendo as colunas Joined e Loan Date End

df = df.drop(['Joined', 'Loan Date End'], axis = 1)
df.head()
```

```python
# Iniciando agora com as colunas 'Value', 'Wage', 'Release Clause'
# Removendo o €

df['Value'] = df['Value'].apply(lambda x: x.replace('€', '').strip())
df['Wage'] = df['Wage'].apply(lambda x: x.replace('€', '').strip())
df['Release Clause'] = df['Release Clause'].apply(lambda x: x.replace('€', '').strip())

df[['Value', 'Wage', 'Release Clause']].head()
```

```python
# Definindo uma função para fazer as multiplicações por 1.000 (K) e 1.000.000 (M)

def converte_din(valor):
  if 'K' in valor:
    k = round((float(valor.replace('K', '').strip()) * 1000), 2)
    return k
  elif 'M' in valor:
    m = round((float(valor.replace('M', '').strip()) * 1000000), 2)
    return m
  else:
    return float(valor)
```

```python
# Convertendo os valores

df['Value'] = df['Value'].apply(converte_din)
df['Wage'] = df['Wage'].apply(converte_din)
df['Release Clause'] = df['Release Clause'].apply(converte_din)

df.head()
```

```python
# Agora vamos remover as estrelas das colunas 'W/F', 'SM' e 'IR'

df['W/F'] = df['W/F'].str[:1]
df['SM'] = df['SM'].str[:1]
df['IR'] = df['IR'].str[:1]

df[['W/F', 'SM', 'IR']].head()
```

```python
# Lá no íncio foi possível observar que a coluna 'Hits' tinha vários dados nulos
# Vamos checar novamente a quantidade

df['Hits'].isnull().sum()
```

```python
# Agora uma observação sobre os dados únicos na coluna

df['Hits'].unique()
```

```python
# Definindo uma função para converter valores

def conversor_valores(valor):
    if pd.notna(valor):
        if isinstance(valor, str) and 'K' in valor:
            return float(valor.replace('K', '')) * 1000
        else:
            return float(valor)
    else:
        return valor
```

```python
# Convertendo os valores

df['Hits'] = df['Hits'].apply(conversor_valores)
df['Hits'].info()
```

```python
# Checando as duplicatas

df.duplicated().sum()
```

```python
# Removendo os dados nulos

df_2 = df.dropna().copy()
df_2.info()
```

```python
# Resetando os index

df_2.reset_index(drop = True, inplace =True)
df_2.info()
```

```python
# Consertando o formato das colunas

df_2['Contract_start_year'] = pd.to_datetime(df_2['Contract_start_year']).dt.year
df_2['Contract_end_year'] = pd.to_datetime(df_2['Contract_end_year']).dt.year
df_2.info()
```

```python
# Salvando o dataset limpo

df_2.to_csv('fifa21_clean.csv', index = False)
```
